```python
# Implementando a função softmax em numpy e pytorch

# Importar bibliotecas
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

#### Usando Numpy

```python
# Fazendo softmax manualmente usando numpy

# Lista de números
z = [1,2,3]

# Computar o resultado do softmax
num = np.exp(z)
den = np.sum( np.exp(z) )

sigma = num / den

print(sigma)
print(np.sum(sigma))
```

```python
# Fazendo a mesma coisa usando números inteiros aleatórios

z = np.random.randint(-5, high = 15, size = 25)
print(z)

# computar o resultado do softmax
num = np.exp(z)
den = np.sum( np.exp(z) )

sigma = num / den

# comparar
plt.plot(z, sigma, 'ko')
plt.xlabel('Original number (z)')
plt.ylabel('Softmax $\sigma$')
plt.title('$\sum\sigma$ = %g' %np.sum(sigma))
plt.show()
```

```python
# comparar
plt.plot(z, sigma, 'ko')
plt.xlabel('Original number (z)')
plt.ylabel('Softmax $\sigma$')
plt.yscale('log')
plt.title('$\sum\sigma$ = %g' %np.sum(sigma))
plt.show()
```

#### Usando Pytorch

```python
# Criando uma instancia do softmax
softfun = nn.Softmax(dim=0)

# Aplicar os dados na função
sigmaT = softfun( torch.Tensor(z) )

# Apresentar os resultados
print(sigmaT)
```

```python
# mostrar que eles são iguais
# comparar
plt.plot(sigma, sigmaT, 'ko')
plt.xlabel('Softmax (numpy)')
plt.ylabel('Softmax (pytorch)')
plt.title(f'Os dois metodos tem correlacao r={np.corrcoef(sigma, sigmaT)[0,1]}')
plt.show()
```
