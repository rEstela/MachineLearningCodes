```python
# importar bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definir o conjunto de pontos para avaliar
x = np.linspace(0.0001, 1, 200)

# computar o log
logx = np.log(x)

# plotar
fig = plt.figure(figsize=(10,4))

# aumentar o tamanho da fonte
plt.rcParams.update({'font.size':15})

plt.plot(x, logx, 'ks-', markerfacecolor = 'w')
plt.xlabel('x')
plt.ylabel('log(x)')
plt.show()
```

```python
# demonstrando que log e exp so inversos

# redefinir com menos pontos
x = np.linspace(0.0001, 1, 20)

# log e exp
logx = np.log(x)
expx = np.exp(x)

# plotar
plt.plot(x, x, color = [.8,.8,.8])
plt.plot(x, np.exp(logx), 'o', markersize=8)
plt.plot(x, np.log(expx), 'x', markersize=8)
plt.xlabel('x')
plt.ylabel('f(g(x))')
plt.legend(['unity', 'exp(log(x)', 'log(exp(x)'])
plt.show()
```
